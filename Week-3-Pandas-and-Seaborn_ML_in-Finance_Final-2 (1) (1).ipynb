```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data =  pd.read_csv('telecom_churn.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
%matplotlib inline
```

```python
data['Area code'].hist();
```

..........................................................

```python
plt.scatter(data['Area code'], data['Total day charge']);
plt.xlabel('Area code');
plt.ylabel('Total day charge');
```

```python
plt.scatter(data['Area code'], data['Total day charge'], color= 'orange');
plt.xlabel('Area code');
plt.ylabel('Total day charge');
```

```python
#Drawing a bicolor diagram

Color =  data['Churn'].map({False: 'blue', True: 'orange'})
Color.head
```

..........................................................

```python
#bicolor diagram

fig= plt.figure(figsize=(11,7))

plt.scatter(data['Area code'], data['Total day charge'], color= Color);
plt.xlabel('Area code');
plt.ylabel('Total day charge');
```

..........................................................

```python

fig= plt.figure(figsize=(11,7))

plt.scatter(data['Area code'], data['Total day charge'], color= Color,
           s= (data['Total day charge']+ 0.05));
plt.xlabel('Area code');
plt.ylabel('Total day charge');
```

```python
pd.crosstab(data['Churn'], data['Area code'])
```
