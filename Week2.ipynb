```python
New_Line = {'age':'25',
            'workclass': 'Private', 'education': 'Some-college','race':'White','sex':'Male'}
print(New_Line)
print(type(New_Line))
```

```python
New_Line_series = pd.Series(New_Line)
print(type(New_Line_series))
print(New_Line_series, '\n')
```

```python
New_Line_dataframe = New_Line_series.to_frame().T
print(type(New_Line_dataframe))
New_Line_dataframe
```

```python
#adding the new line dataframe to data dataframe end
dataset = pd.concat([dataset, New_Line_dataframe], ignore_index='true')
dataset
```

```python
# Grouping "relationship" and "hours-per-week" (original)

grouped_original = dataset.groupby(['relationship', 'hours-per-week']).size()
display(grouped_original)
```

```python
# Function
n=2
def reduce_hours(row):
    if pd.notna(row['hours-per-week']):
        row['hours-per-week'] = row['hours-per-week']- n
    return row
```

```python
dataset_reduced = dataset.apply(reduce_hours, axis=1)
```

```python
display(dataset_reduced)
```

```python
# Grouping "relationship" and reduced 'hours-per-week'

grouped_reduced = dataset_reduced.groupby(['relationship', 'hours-per-week']).size()
display(grouped_reduced)
```
